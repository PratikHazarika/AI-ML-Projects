# User playlist information
The information used to create custom playlist of the user

```python
user_playlist_name = ''
user_playlist_songs = []
total_songs = 0
```

```python
CATEGORIES = None # categories used to build custom playlist
mood = get_mood() # sentiment analysis on user input to detect the mood from main.py
```

# Updating categories based on user's mood

```python
if mood == 'happy':
    CATEGORIES = HAPPY_CATEGORIES
    user_playlist_name = "Happy" 

if mood == 'party':
    CATEGORIES = PARTY_CATEGORIES
    user_playlist_name = "Party"

if mood == 'sad':
    CATEGORIES = SAD_CATEGORIES
    user_playlist_name = "Sad"

if mood == 'chill':
    CATEGORIES = CHILL_CATEGORIES + RAP_CATOGRIES + SAD_CATEGORIES
    user_playlist_name = "Chill"

if mood == 'neutral':
    CATEGORIES = HAPPY_CATEGORIES + RAP_CATOGRIES
    user_playlist_name = "Neutral"

CATEGORIES = set(CATEGORIES)
```

# Finding songs in Spotify
Using the Spotify API to surf through songs on Spotify to find the appropriate songs to build a playlist based on the user's mood

```python
for category in CATEGORIES:
    
  if total_songs < 99:

      playlist_names = spotify.category_playlists(category_id=category, country='US', limit=5)['playlists']['items']

      for playlist_name in playlist_names:
          numSongs = 0

          #id of that playlist
          id = spotify.search(q = playlist_name['name'], limit=1, type='playlist')['playlists']['items'][0]['id']

          #the songs of the playlist
          playlist_songs = spotify.playlist(playlist_id=id, fields='tracks', market="US")['tracks']

          #printing top 5 songs in the playlist
          for i in range(len(playlist_songs)):
              if numSongs < 5:
                  track = playlist_songs['items'][i]['track']
                  
                  if track is not None:
                      user_playlist_songs.append(track['uri'])
                      total_songs += 1
                      numSongs += 1
              
              else:
                  break
      else:
          break
```
